```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./HeartDisease/Cleavland.csv' , header=None)
df.head()
```

```python
df.columns=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
df.head()
```

```python
df.isnull().sum().sum()
```

```python
(df.values=='?').sum()
```

```python
df=df.replace('?',np.nan)
df=df.dropna()
```

```python
(df.values=='?').sum()
```

```python
(df.values==np.NaN).sum()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df['age'],bins=10,color='skyblue')
plt.xlabel('age')
plt.ylabel('freq')
plt.title('age vs freq')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['sex'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['lightblue', 'lightpink'])
plt.ylabel('')
plt.title('sex distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='cp',y='chol',data=df)
plt.xlabel('chest pain type')
plt.ylabel('cholestrol')
plt.title('chest pain vs cholestrol')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age',y='thalach',data=df,hue='num')
plt.xlabel('age')
plt.ylabel('heart rate')
plt.title('age vs heart rate')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['num'].value_counts().plot(kind='bar',color='lightgreen')
plt.xlabel('heart disease')
plt.ylabel('freq')
plt.title('heart disease distribution')
plt.show()
```

```python
fig,axes =plt.subplots(1,3,figsize=(10,5))
sns.boxplot(x='chol',data=df,ax=axes[0])
axes[0].set_title('Box Plot')
sns.histplot(data=df['chol'],ax=axes[1])
axes[1].set_title('Histogrm')
sns.scatterplot(x='chol',y='thal',data=df,ax=axes[2])
axes[2].set_title('scatter plot')

plt.tight_layout()
plt.show()
```
