```python
data=df
```

```python
plt.figure(figsize=(8,6))
plt.hist(data['age'],color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(data['education-num'],'o', color='lightpink')
plt.xlabel('Index')
plt.ylabel('Education Number')
plt.title('Distribution of Education Number')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['workclass'].value_counts().plot(kind='bar',color='lightgreen')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Distribution of Workclass')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(data['hours-per-week'],color='purple')
plt.xlabel('Index')
plt.ylabel('Hours per Week')
plt.title('Hours per Week Trend')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='age',y='education-num',data=df,hue='race')
plt.xlabel('Age')
plt.ylabel('Education Number')
plt.title('Age vs Education Number')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['race'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.ylabel('')
plt.title('Race Composition')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='native-country',y='capital-gain',data=df)
plt.xlabel('native-country')
plt.ylabel('Capital Gain')
plt.title('Distribution of Capital Gain')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='age',y='hours-per-week',data=df,hue='race')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.title('Age vs Hours per Week')
plt.show()
```

```python
fig,axes =plt.subplots(1,3,figsize=(10,5))
sns.boxplot(x='chol',data=df,ax=axes[0])
axes[0].set_title('Box Plot')
sns.histplot(data=df['chol'],ax=axes[1])
axes[1].set_title('Histogrm')
sns.scatterplot(x='chol',y='thal',data=df,ax=axes[2])
axes[2].set_title('scatter plot')

plt.tight_layout()
plt.show()
```
